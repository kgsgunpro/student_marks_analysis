```python
import pandas as pd
import numpy as np

# Generate random names
student_names = [
    'Alice',
    'Bob',
    'Charlie',
    'David',
    'Eve',
    'Frank',
    'Grace',
    'Heidi',
    'Ivan',
    'Judy'
]

# Generate random marks for Math, Science, and English
np.random.seed(42) # for reproducibility
math_marks = np.random.randint(40, 100, size=len(student_names))
science_marks = np.random.randint(35, 95, size=len(student_names))
english_marks = np.random.randint(50, 100, size=len(student_names))

# Create a dictionary for the dataset
student_data = {
    'Student Name': student_names,
    'Math Score': math_marks,
    'Science Score': science_marks,
    'English Score': english_marks
}

# Create a Pandas DataFrame
df_students = pd.DataFrame(student_data)

print("Generated Student Performance DataFrame:")
print(df_students.head())
```

```python
# Calculate each student's total score and add it as a new column
df_students['Total Score'] = df_students['Math Score'] + df_students['Science Score'] + df_students['English Score']

print("DataFrame with 'Total Score' column:")
print(df_students.head())
```

```python
# Calculate each student's average score and add it as a new column
df_students['Average Marks'] = df_students[['Math Score', 'Science Score', 'English Score']].mean(axis=1)

print("DataFrame with 'Average Marks' column:")
print(df_students.head())
```

```python
# Calculate average scores for each subject
average_math_score = df_students['Math Score'].mean()
average_science_score = df_students['Science Score'].mean()
average_english_score = df_students['English Score'].mean()

print(f"Average Math Score: {average_math_score:.2f}")
print(f"Average Science Score: {average_science_score:.2f}")
print(f"Average English Score: {average_english_score:.2f}")
```

```python
import matplotlib.pyplot as plt

# Create a DataFrame for average scores for easier plotting
average_scores_df = pd.DataFrame({
    'Subject': ['Math', 'Science', 'English'],
    'Average Score': [average_math_score, average_science_score, average_english_score]
})

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(average_scores_df['Subject'], average_scores_df['Average Score'], color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.title('Average Scores by Subject')
plt.ylim(0, 100) # Assuming scores are out of 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
std_dev_math = df_students['Math Score'].std()
std_dev_science = df_students['Science Score'].std()
std_dev_english = df_students['English Score'].std()

print(f"Standard Deviation of Math Scores: {std_dev_math:.2f}")
print(f"Standard Deviation of Science Scores: {std_dev_science:.2f}")
print(f"Standard Deviation of English Scores: {std_dev_english:.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_students['Student Name'], df_students['Total Score'], color='purple')
plt.xlabel('Student Name')
plt.ylabel('Total Score')
plt.title('Total Scores per Student')
plt.xticks(rotation=45, ha='right') # Rotate names for better readability if they overlap
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_students['Student Name'], df_students['Average Marks'], color='teal')
plt.xlabel('Student Name')
plt.ylabel('Average Marks')
plt.title('Average Marks per Student')
plt.xticks(rotation=45, ha='right') # Rotate names for better readability if they overlap
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by 'Average Marks' in descending order and get the top 3 students
top_3_students = df_students.sort_values(by='Average Marks', ascending=False).head(3)

print("Top 3 Students based on Average Marks:")
print(top_3_students[['Student Name', 'Average Marks']])
```
